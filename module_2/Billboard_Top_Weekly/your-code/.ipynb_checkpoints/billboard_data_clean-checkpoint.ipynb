```python
import pandas as pd
import numpy as np
```

1. Import of original data

```python
billboard_100 = pd.read_csv("/BillboardFromLast20/billboardHot100_1999-2019.csv")
```

```python
billboard_100.head()
```

2. Genre column has a string with different genres separated by a comma. We will create a list and then obtain the top genre, which will be the first element on the list.

```python
billboard_100["Genre"] = billboard_100["Genre"].str.split(",")
```

```python
billboard_100.head()
```

```python
billboard_100["Genre"][0][0]
```

```python
genres = list(billboard_100["Genre"])
```

```python
top_genres = [i[0] for i in genres]
```

```python
billboard_100["top_genre"] = top_genres
```

```python
billboard_100.head(10)
```

```python
billboard_100["artists_original"] = billboard_100["Artists"]
```

3. It was analyzed that one of the Artists had a comma on it's name so the comma was removed in order to create the lists of artists. As well as the Genre column, the main artist was considered the first on the list.

```python
billboard_100["Artists"] = (np.where(billboard_100["Artists"] == "Tyler, The Creator"
                                     , "Tyler The Creator"
                                     , billboard_100["Artists"]))
```

```python
billboard_100.loc[np.where(billboard_100["Artists"] == "Tyler, The Creator")]
```

```python
billboard_100.head()
```

```python
billboard_100["Artists"] = billboard_100["Artists"].str.split(",")
```

```python
artists = list(billboard_100["Artists"])
```

```python
billboard_100["main_artist"] = [i[0] for i in artists]
```

```python
billboard_100.head()
```

4. In order to find out if the song was performed by one or more artists, the len of the lists created form artists was used to determine how many artists performed in a song.

```python
len(billboard_100["Artists"][3])
```

```python
#billboard_100 = billboard_100.drop(columns="Unnamed: 0")
```

```python
billboard_100.head()
```

```python
billboard_100["no_artists"] = [len(billboard_100["Artists"][i]) for i in range(len(billboard_100["Artists"]))]
```
